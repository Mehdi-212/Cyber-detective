```python
import pyodbc
import pandas as pd
from pandasql import sqldf
```

```python
df_books = pd.read_csv('books.csv')

books = df_books[['index', 'titre', 'prix', 'etoile', 'dispo', 'link']]
query = "SELECT * FROM books"
books = sqldf(query, globals())
books
```

```python
conn = pyodbc.connect('Driver={SQL Server};'
                      'Server=UTILISA-REVMD1C\SQLEXPRESS;'
                      'Database=books;'
                      'Trusted_Connection=yes;')

cursor = conn.cursor()
sql = "INSERT INTO book(id, titre, prix, etoile, dispo, link) VALUES(?,?,?,?,?,?)"
for i,row in books.iterrows():
    cursor.execute(sql, tuple(row))    
    conn.commit()
```
