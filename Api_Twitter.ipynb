```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
# Création d'une liste pour ajouter les données de tweets
liste_tweets = []

# Utilisation de TwitterSearchScraper pour récupérer les données et ajouter les tweets à la listefor i,tweet in enumerate(sntwitter.TwitterSearchScraper('the Art of War since:2022-06-01 until:2023-01-01').get_items()):
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('the Art of War since:2022-06-01 until:2023-01-01').get_items()):
    if i>1000:
        break
    liste_tweets.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
 
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('The song of Achille since:2022-06-01 until:2023-01-01').get_items()):
    if i>1000:
        break
    liste_tweets.append([tweet.date, tweet.id, tweet.content, tweet.user.username])

for i,tweet in enumerate(sntwitter.TwitterSearchScraper('Batman : the Dark Night Return since:2022-06-01 until:2023-01-01').get_items()):
    if i>1000:
        break    
    liste_tweets.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
    
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('The Picture of Dorian Gray since:2022-06-01 until:2023-01-01').get_items()):
    if i>1000:
        break
    liste_tweets.append([tweet.date, tweet.id, tweet.content, tweet.user.username])

for i,tweet in enumerate(sntwitter.TwitterSearchScraper('The Book Thief since:2022-06-01 until:2023-01-01').get_items()):
    if i>1000:
        break
    liste_tweets.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
    
# Création d'un dataframe à partir de la liste de tweets ci-dessus

df_tweets = pd.DataFrame(liste_tweets, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
df_tweets
```

```python
#nettoyage des données (utilisation de la librairie regex)
df_tweets['Text'] = df_tweets['Text'].str.replace(r"[A-Za-z.][0-9]+[A-Za-z%°.]", " ")
df_tweets['Text'] = df_tweets['Text'].str.replace(r"(\s-\s|-$)", " ")
df_tweets['Text'] = df_tweets['Text'].str.replace(r"[,!?%()/]", " ")
df_tweets['Text'] = df_tweets['Text'].str.replace(r"&\S*\s", " ")
df_tweets['Text'] = df_tweets['Text'].str.replace(r"&", " ")
df_tweets['Text'] = df_tweets['Text'].str.replace(r"+", " ")
df_tweets['Text'] = df_tweets['Text'].str.replace(r"#", " ")
df_tweets['Text'] = df_tweets['Text'].str.replace(r"$", " ")
df_tweets['Text'] = df_tweets['Text'].str.replace(r"£", " ")
df_tweets['Text'] = df_tweets['Text'].str.replace(r"%", " ")                                                  
df_tweets['Text'] = df_tweets['Text'].str.replace(r"@", " ")
df_tweets['Text'] = df_tweets['Text'].str.replace(r"-", " ")

df_tweets
```

```python
# transformation du dataframe en CSV
df_tweets.to_csv('tweets.csv', index=False)
```
